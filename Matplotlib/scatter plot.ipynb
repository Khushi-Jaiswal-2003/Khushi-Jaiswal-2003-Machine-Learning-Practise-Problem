```python
df_google = pd.read_csv("googleplaystore.csv", nrows=1000)
```

```python
x = df_google['Rating']
y = df_google["Reviews"]

plt.scatter(x,y)
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(16,9))
x = df_google['Rating']
y = df_google["Reviews"]

plt.scatter(x,y, c='r', marker='*', s= 100, alpha=0.5, linewidth=7, edgecolors='g')
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(16,9))
x = df_google['Rating']
y = df_google["Reviews"]

plt.scatter(x,y, c='r', marker='*', s= 100, alpha=0.5, linewidth=7, edgecolors='g')

plt.scatter(x,df_google['Installs'], c='k', marker='o', s= 100, alpha=0.5, linewidth=7, edgecolors='r')

plt.title("Google Play Store Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review & Installs")
plt.show()
```
