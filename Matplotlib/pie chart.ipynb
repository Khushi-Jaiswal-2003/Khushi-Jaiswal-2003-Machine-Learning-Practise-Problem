```python
from matplotlib import pyplot as plt
```

```python
classes = ["Pyhton", "R", "Machine Learning", "Artifical Intelligence", "Data Science"]
class1_students = [45,15,35,25,30]
plt.pie([1])
plt.show()
```

```python
classes = ["Pyhton", "R", "Machine Learning", "Artifical Intelligence", "Data Science"]
class1_students = [45,15,35,25,30]
plt.pie(class1_students, labels=classes)
plt.show()
```

```python
classes = ["Pyhton", "R", "Machine Learning", "Artifical Intelligence", "Data Science"]
explode=[0.03,0,0.1,0,0]
color = ['c','b','r','y','g']
text ={'fontsize':15}
plt.pie(class1_students, labels=classes, explode=explode, colors=color, autopct='%0.2f%%', shadow=True, radius=1.4, startangle=90, textprops=text)
plt.show()
```

```python
plt.figure(figsize=(15,5))
classes = ["Pyhton", "R", "Machine Learning", "Artifical Intelligence", "Data Science"]
explode=[0.03,0,0.1,0,0]
color = ['c','b','r','y','g']
text ={'fontsize':15}
wedge={"linewidth":3, 'width':3, "edgecolor":"k"}
plt.pie(class1_students, labels=classes, explode=explode, colors=color, autopct='%0.2f%%', pctdistance=0.6,labeldistance=1.1, shadow=False, radius=1.4, startangle=90,counterclock=False,wedgeprops=wedge,frame=False, textprops=text,rotatelabels=True)
plt.show()
```

```python
classes = ["Pyhton", "R", "Machine Learning", "Artifical Intelligence", "Data Science"]
explode=[0.03,0,0.1,0,0]
color = ['c','b','r','y','g']
text ={'fontsize':15}
plt.pie(class1_students, labels=classes, explode=explode, colors=color, autopct='%0.2f%%', shadow=True, radius=1.4, startangle=90, textprops=text)
plt.legend(loc =5)
plt.show()
```
