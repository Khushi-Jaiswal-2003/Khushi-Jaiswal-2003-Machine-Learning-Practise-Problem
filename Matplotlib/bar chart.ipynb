```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]

plt.bar(classes, class1_students)
```

```python
plt.barh(classes, class1_students)
```

```python
plt.bar(classes, class1_students, width=0.9, align="edge", color = 'y', edgecolor = 'r', linewidth = 4, alpha =0.5, linestyle ='-', label="Class_1 students", visible = True)

plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(10,6))
plt.bar(classes, class1_students, width=0.9, align="edge", color = 'y', edgecolor = 'k', linewidth = 4, linestyle ='-', label="Class_1 students", visible = True)

plt.title("Bar Chart Of class", fontsize = 15)
plt.xlabel("Programming Languages", fontsize = 13)
plt.ylabel("No. Of Students", fontsize =13)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
# plt.figure(figsize=(10,6))
plt.bar(classes, class1_students, width=0.5, align="edge", color = 'y', edgecolor = 'k', linewidth = 4, linestyle ='-', label="Class_1 students", visible = True)

plt.bar(classes, class2_students, width=0.5, align="edge", color = 'g', edgecolor = 'k', linewidth = 4, linestyle ='-', label="Class_2 students", visible = True)

plt.title("Bar Chart Of class", fontsize = 15)
plt.xlabel("Programming Languages", fontsize = 13)
plt.ylabel("No. Of Students", fontsize =13)
plt.legend()
plt.show()
```

```python
class_index = np.arange(len(classes))

width=0.3

plt.bar(class_index, class1_students, width=width, color = 'y', edgecolor = 'k', label="Class_1 students")

plt.bar(class_index+width, class2_students, width=width, color = 'g', edgecolor = 'k', label="Class_2 students")

plt.bar(class_index+width+width, class2_students, width=width, color = 'r', edgecolor = 'k', label="Class_3 students")

plt.xticks(class_index+width, classes, rotation=25)

plt.title("Bar Chart Of class", fontsize = 15)
plt.xlabel("Programming Languages", fontsize = 13)
plt.ylabel("No. Of Students", fontsize =13)
plt.legend()
plt.show()
```
