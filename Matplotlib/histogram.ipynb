```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
ml_students_age = np.random.randint(18,45,(100))
py_students_age = np.random.randint(18,50,(100))
```

```python
plt.hist(ml_students_age, bins=[15,20,25,30,35,40,45,50],rwidth=0.8, orientation='vertical', color='y', label="ML students")

plt.title("Maching Learning Students Data")
plt.xlabel("Students Age")
plt.ylabel("No.Students Age")
plt.legend()
plt.show()
```

```python
plt.hist([ml_students_age,py_students_age], bins=[15,20,25,30,35,40,45],rwidth=0.8, histtype='bar', align='mid', orientation='vertical', color=['y','r'], label=["ML students",'Py students'])

plt.title("Maching Learning and Python Students Data")
plt.xlabel("Students Age")
plt.ylabel("No.Students Age")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16,9))

plt.hist([ml_students_age,py_students_age], bins=[15,20,25,30,35,40,45],rwidth=0.8, histtype='bar', align='mid', orientation='vertical', color=['y','r'], label=["ML students","Py students"])
plt.title("Maching Learning and Python Students Data")
plt.xlabel("Students Age")
plt.ylabel("No.Students Age")
plt.legend()
plt.show()
```

```python
from matplotlib import style

style.use('ggplot')
plt.figure(figsize = (16,9))

plt.hist([ml_students_age,py_students_age], bins=[15,20,25,30,35,40,45],rwidth=0.8, histtype='bar', align='mid', orientation='vertical', color=['y','r'], label=["ML students","Py students"])
plt.title("Maching Learning and Python Students Data")
plt.xlabel("Students Age")
plt.ylabel("No.Students Age")
plt.legend()
plt.show()
```
